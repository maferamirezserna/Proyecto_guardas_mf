```python
import sqlite3

def check_table_structure(db_file):
    connection = sqlite3.connect(db_file)
    cursor = connection.cursor()
    
    # Obtener la información de la tabla
    cursor.execute("PRAGMA table_info(empleados)")
    columns = cursor.fetchall()
    
    # Mostrar la estructura de la tabla
    print("Columnas en la tabla 'empleados':")
    for column in columns:
        print(column)
    
    connection.close()

# Nombre del archivo de base de datos
db_file = 'mi_base_de_datos.db'

# Verificar la estructura de la tabla
check_table_structure(db_file)
```

```python
import sqlite3

# Consulta de la Base de datos en SQL

def open_connection(db_file):
    """ Abre una conexión a la base de datos SQLite """
    conn = sqlite3.connect(db_file)
    return conn

def close_connection(conn):
    """ Cierra la conexión a la base de datos """
    if conn:
        conn.close()

def fetch_data(conn, table_name):
    """ Recupera todos los datos de una tabla y los muestra """
    try:
        cursor = conn.cursor()
        query = f"SELECT * FROM \"{table_name}\""
        cursor.execute(query)
        rows = cursor.fetchall()
        
        # Mostrar los datos
        if rows:
            for row in rows:
                print(row)
        else:
            print(f"No hay datos en la tabla '{table_name}'.")
    
    except sqlite3.Error as e:
        print(f"Error al recuperar datos: {e}")

def main():
    db_file = 'mi_base_de_datos.db'
    table_name = 'empleados'  # Cambia el nombre de la tabla según tu base de datos
    
    # Abrir la conexión
    conn = open_connection(db_file)
    
    try:
        # Recuperar y mostrar datos
        fetch_data(conn, table_name)
    
    finally:
        # Cerrar la conexión
        close_connection(conn)

if __name__ == '__main__':
    main()
```
